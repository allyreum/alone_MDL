# **K-NN을 이용한 알고리즘**

````python
length = bream_length + smelt_length
weight = bream_weight + smelt_weight # 리스트 합치는 방법
#print(length)
#print(weight)
````

**사이킷 런(머신러닝 패키지)을 이용하기 위해선 2차원 리스트 형태의 데이터가 필요함**

````python
fish_data = [[l,w] for l, w in zip(length, weight)]
# zip 함수는 나열된 리스트 각각 하나씩 원소를 꺼내 반환

print(fish_data[:5])
````

````python
fish_target = [1] * 35 + [0] * 14 # 결과값 생성: 1 = 도미, 0 = 빙어

print(fish_target)
````

K-NN 알고리즘을 구현한 클래스인 **KNeighborsClassifier** import

````python
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()
````

````python
kn.fit(fish_data, fish_target)      # trainning -> fit(): 주어진 데이터로 알고리즘 훈련
kn.score(fish_data, fish_target)    # model 평가(Accuracy)
````

(30, 600)의 생선을 분류하면?

````python
kn.predict([[30,600]]) # predict은 새로운 데이터의 정답을 예측
````

K-NN

장점: 새로운 데이터 예측 시엔 가장 가까운 직선거리의 데이터를 보게됨.

단점: 데이터 양이 큰 경우, 직선거리 산출 시 많은 시간 소요되고, 메모리로 많이 필요함.

````python
print(kn._fit_X)    # fish data를 모두 갖고 있음
print(kn._y)
````

**KNeighborsClassifier클래스는 기본적으로 가까운 5개의 데이터를 참고한다. (K = 5)**
하지만 **n_neighbors** 매개변수를 통해 변경이 가능하다.

````python
kn49 = KNeighborsClassifier(n_neighbors = 49)   # n_neighbors를 통해 매개변수 변경가능, k = 49개로 한 모델

kn49.fit(fish_data, fish_target)
kn49.score(fish_data, fish_target)
````

accuracy가 1에서 0.71로 낮아진 걸로 보았을 때 k=5로 보는 것이 도미와 빙어를 분류하는데 좋은 모델이었던 것으로 보인다!

fish data는 도미가 35개, 빙어가 14개로 구성되어 있음.

kn49는 49개 중 도미 35개로 다수를 차지 하니까 어떤 데이터를 넣어도 무조건 도미를 예측할 것임.

````python
print(35/49)    # 정확도 = 맞춘 개수 / 전체 개수, kn49는 도미만 올바르게 맞치기 때문에 score() 매서드와 같은 값을 얻을 수 있음.
````

# 확인문제
5 <= k <= 49 중에 score < 1로 되는 k를 구하여라

````python
kn = KNeighborsClassifier()
kn.fit(fish_data, fish_target)

for n in range(5, 50):
  kn.n_neighbors = n # 인접 이웃 개수 설정
  score = kn.score(fish_data, fish_target) # 점수 개산
  if score < 1:
    print(n, score)
    break
````

## 3-2. 선형회귀

````python
import numpy as np
# insert data(perch_length, perch_weight)

from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(perch_length, perch_weight, random_state = 42)

train_input = train_input.reshape(-1,1) # sklearn에서 2차원 배열 만들어야함!!!
test_input = test_input.reshape(-1,1)

from sklearn.neighbors import KNeighborsRegressor
knr = KNeighborsRegressor(n_neighbors = 3)
knr.fit(train_input, train_target)

#50cm인 농어의 무게 예측
print(knr.predict([[50]]))    # knr.predict([[50]]) (2차원 표기)
````

실제로 더 많은 무게가 나간다고 했을 때, 예측이 잘못된 이유는?

````python
import matplotlib.pyplot as plt
distances, indexes = knr.kneighbors([[50]]) # 50cm 농어의 이웃
# distances: 50cm 농어와 가장 가까운 이웃들과의 거리 (유클리드 거리)
# indexes: 그 이웃들이 훈련 데이터에서 몇 번째 샘플이었는지 알려주는 인덱스

plt.scatter(train_input, train_target)
plt.scatter(train_input[indexes], train_target[indexes], marker = 'D')
plt.scatter(50, knr.predict([[50]]), marker='^')  # 예측값 표시
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
````

````python
print(np.mean(train_target[indexes]))
print(knr.predict([[100]]))
print(knr.predict([[50]]))
````

50cm, 100cm 농어의 무게가 동일함

````python
distances, indexes = knr.kneighbors([[100]])
plt.scatter(train_input, train_target)
plt.scatter(train_input[indexes], train_target[indexes], marker = 'D')
plt.scatter(100, knr.predict([[100]]), marker='^')  # 예측값 표시
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
````

근접한 이웃을 기반으로 하기 때문에 농어의 길이가 길어도 무게는 늘어나지 않음.

따라서 **회귀**를 이용해서 구해야하 함!!!

**sklearn.linear_model** 패키지 아래에 **LinearRegression** 클래스로 선형회귀 알고리즘 구현 가능

````python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

lr.fit(train_input, train_target)
print(lr.coef_, lr.intercept_) #(기울기, 절편) -> model parameter에 해당함
print(lr.predict([[50]]))
````

훈련 > 테스트 : 과대적합

그래프를 보았을 때 직선으로는 만족하지 않아 보임. 곡선인 **다항회귀(다항식을 이용한 선형회귀)**로 모델을 만들어야 함

**2차 방정식**은 길이를 **제곱 항**이 훈련set에 추가해야 함

→ x²과 x를 각각의 특성(feature) 으로 넣어줘야 하고, **np.column_stack((x², x))** 이렇게 2개의 열로 만들어줌

````python
train_poly = np.column_stack((train_input**2, train_input))
test_poly = np.column_stack((test_input**2, test_input))
print(train_poly.shape, test_poly.shape)
````

````python
lr = LinearRegression()
lr.fit(train_poly, train_target)
print(lr.predict([[50**2, 50]]))
print(lr.coef_, lr.intercept_)  # 무게 = 1.01 x 길이² - 21.6 x 길이 + 116.1
````

````python
point = np.arange(15, 50) # 15 ~ 49 출력
plt.scatter(train_input, train_target)
plt.plot(point, 1.01 * point**2 - 21.6 * point + 116.1)
plt.scatter(50, 1574, marker = '^')
plt.show()
````

````python
print(lr.score(train_poly, train_target))
print(lr.score(test_poly, test_target))
````

## 3-3. 특성공학과 규제: **릿지와 라쏘**

- 특성공학? 기존의 특성을 사용해 새로운 특성을 뽑아내는 것(ex. 농어길이 x 농어높이)


````python
import pandas as pd
import numpy as np
df = pd.read_csv('https://bit.ly/perch_csv_data')
perch_full = df.to_numpy()
# csv파일 → 판더스 df(pd.read_csv()) → 넘파일 배열(to_numpy())
perch_weight = np.array([5.9, 32.0, 40.0, 51.5, 70.0, 100.0, 78.0, 80.0, 85.0, 85.0, 110.0,
       115.0, 125.0, 130.0, 120.0, 120.0, 130.0, 135.0, 110.0, 130.0,
       150.0, 145.0, 150.0, 170.0, 225.0, 145.0, 188.0, 180.0, 197.0,
       218.0, 300.0, 260.0, 265.0, 250.0, 250.0, 300.0, 320.0, 514.0,
       556.0, 840.0, 685.0, 700.0, 700.0, 690.0, 900.0, 650.0, 820.0,
       850.0, 900.0, 1015.0, 820.0, 1100.0, 1000.0, 1100.0, 1000.0,
       1000.0])

from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(perch_full, perch_weight, random_state=42)
````

사이킷런에는 특성을 만들거나 전처리 가능한 class가 존재함

````python
# PolynomialFeatures(객체 생성 > fit > transform)
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures()     # 제곱항, 특성을 곱한 항 추가, default = 1(절편)
poly.fit([[2,3]])
print(poly.transform([[2,3]]))
````

````python
poly = PolynomialFeatures(include_bias = False) # include_biase = F: default = 1 을 제외
poly.fit([[2,3]])
print(poly.transform([[2,3]]))
````

````python
poly = PolynomialFeatures(include_bias = False)
poly.fit(train_input)
train_poly = poly.transform(train_input)
test_poly = poly.transform(test_input)

print(train_poly.shape, test_poly.shape)
poly.get_feature_names_out()
````

# **다중회귀 모델 훈련하기**

````python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(train_poly, train_target)
print(lr.score(train_poly, train_target))
print(lr.score(test_poly, test_target))
````

고차항 추가(5제곱)

````python
poly = PolynomialFeatures(degree = 5, include_bias =False)
poly.fit(train_input)
train_poly = poly.transform(train_input)
test_poly = poly.transform(test_input)
print(train_poly.shape)
lr.fit(train_poly, train_target)
print(lr.score(train_poly, train_target))
print(lr.score(test_poly, test_target))     # 특성개수를 너무 늘리면, 과대적합하게 됨,,
````

**규제** : ML 모델이 훈련세트를 과도하게 학습하지 못하도록 훼방 놓는 것을 의미함(계수의 크기를 작게 만듦)

**StandardScaler** 클래스는 변환기 중 하나이며, **train set으로 학습한 변환기는 test set도 적용**해야함

````python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(train_poly)
train_scaled = ss.transform(train_poly)
test_scaled = ss.transform(test_poly)
````

- 릿지는 계수를 제곱한 값으로 규제를 적용
- 라쏘는 계수의 절댓값을 기준으로 규제를 적용, 0으로 만들 수 있음

# **릿지회귀**

````python
from sklearn.linear_model import Ridge
ridge = Ridge()
ridge.fit(train_scaled, train_target)
print(ridge.score(train_scaled, train_target))
print(ridge.score(test_scaled, test_target))
````

적절한 alpha값을 찾는 방법은 alpha에 대한 R² 그래프를 그려보는 것


````python
import matplotlib.pyplot as plt
train_score = []
test_score = []

alpha_list = [0.001, 0.01, 0.1, 1, 10, 100]
for alpha in alpha_list:
  ridge = Ridge(alpha = alpha)
  ridge.fit(train_scaled, train_target)
  train_score.append(ridge.score(train_scaled, train_target))
  test_score.append(ridge.score(test_scaled, test_target))

plt.plot(np.log10(alpha_list), train_score)
plt.plot(np.log10(alpha_list), test_score)
plt.show()
# 그래프 중 위는 train, 아래는 test
````

````python
ridge = Ridge(alpha = 0.1)
ridge.fit(train_scaled, train_target)
print(ridge.score(train_scaled, train_target))
print(ridge.score(test_scaled, test_target))
````

# **라쏘회귀**

````python
from sklearn.linear_model import Lasso
lasso = Lasso()
lasso.fit(train_scaled, train_target)
print(lasso.score(train_scaled, train_target))
print(lasso.score(test_scaled, test_target))

import matplotlib.pyplot as plt
train_score = []
test_score = []

alpha_list = [0.001, 0.01, 0.1, 1, 10, 100]
for alpha in alpha_list:
  lasso = Lasso(alpha = alpha)
  lasso.fit(train_scaled, train_target)
  train_score.append(lasso.score(train_scaled, train_target))
  test_score.append(lasso.score(test_scaled, test_target))

plt.plot(np.log10(alpha_list), train_score)
plt.plot(np.log10(alpha_list), test_score)
plt.show()
# 그래프 중 위는 train, 아래는 test
````

````python
lasso = Lasso(alpha = 10)
lasso.fit(train_scaled, train_target)
print(lasso.score(train_scaled, train_target))
print(lasso.score(test_scaled, test_target))
````

# **4장. 다양한 분류 알고리즘**

**4-1. 로지스틱 회귀**

````python
import pandas as pd
fish = pd.read_csv('https://bit.ly/fish_csv_data')
print(pd.unique(fish['Species'])) # fish[['Species']]하면 fish_target이 2차원 배열이 됨
````

````python
print(fish.columns)

fish_input = fish[['Weight', 'Length', 'Diagonal', 'Height', 'Width']].to_numpy() # 넘파이 배열로 변경
print(fish_input[:5])
````

````python
fish_target = fish['Species'].to_numpy()

from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(fish_input, fish_target, random_state = 42)

from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
````

K-Neighbor 분류기의 확률예측

````python
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier(n_neighbors = 3)
kn.fit(train_scaled, train_target)
print(kn.score(train_scaled, train_target))
print(kn.score(test_scaled, test_target))
````

````python
print(kn.classes_)                  # 자동적으로 알파벳순서로 지정
print(kn.predict(test_scaled[:5]))  # 5개의 sample만 선택

import numpy as np
proba = kn.predict_proba(test_scaled[:5])     # 모델이 예측할 때 "각 클래스에 속할 확률" 을 알려주는 method, probability를 표기
print(np.round(proba, decimals = 2))          # decimals = 2 소수점 둘째자리까지 표기
````

````python
distances, indexes = kn.kneighbors(test_scaled[3:4])  # 3번째 indexing & 2차원배열
print(train_target[indexes])                          # 이런 걸 간단히 하는 게 predict_proba
````

### 로지스틱 회귀로 이진 분류 수행하기

````python
char_arr = np.array(['A','B','C','D','E'])
print(char_arr[[True, False, True, False, False]])

bream_smelt_indexes = (train_target == 'Bream') | (train_target == 'Smelt')
train_bream_smelt = train_scaled[bream_smelt_indexes]
target_bream_smelt = train_target[bream_smelt_indexes]

from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(train_bream_smelt, target_bream_smelt)
````

````python
print(lr.predict(train_bream_smelt[:5]))

print(lr.classes_)
print(lr.predict_proba(train_bream_smelt[:5]))
````

````python
print(lr.coef_, lr.intercept_)

decisions = lr.decision_function(train_bream_smelt[:5])
print(decisions)

# decision_function은 모델이 계산한 "결정 점수"
# 값이 0보다 크면 → 'Bream' (Positive Class)로 예측
# 값이 0보다 작으면 → 'Smelt' (Negative Class)로 예측

from scipy.special import expit
# expit() 함수는 시그모이드 함수(=로지스틱 함수) 계산, 시그모이드는 아무 숫자든 0~1 사이로 변환
print(expit(decisions))
````

### 로지스틱 회귀로 다중 분류 수행하기

````python
lr = LogisticRegression(C=20, max_iter = 1000)
lr.fit(train_scaled, train_target)
print(lr.score(train_scaled, train_target))
print(lr.score(test_scaled, test_target))

print(lr.predict(test_scaled[:5]))
````

````python
proba = lr.predict_proba(test_scaled[:5])
print(np.round(proba, decimals = 3))
print(lr.classes_)
print(lr.coef_.shape, lr.intercept_.shape)
````

````python
decision = lr.decision_function(test_scaled[:5])
print(np.round(decision, decimals = 3))

from scipy.special import softmax
proba = softmax(decision, axis = 1)
print(np.round(proba, decimals = 3))
````
