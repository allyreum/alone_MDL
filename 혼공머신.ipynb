````python
bream_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0,
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0,
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0]
bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0,
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0,
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0]
````

````python
import matplotlib.pyplot as plt

plt.scatter(bream_length, bream_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
````

````python
smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]
````

````python
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
````

````python
length = bream_length + smelt_length
weight = bream_weight + smelt_weight # 리스트 합치는 방법
print(length)
print(weight)
````

**사이킷 런을 이용하기 위해선 2차원 리스트 형태의 데이터가 필요함**

````python
fish_data = [[l,w] for l, w in zip(length, weight)]
# zip 함수는 나열된 리스트 각각 하나씨구원소를 꺼내 반환

print(fish_data)
````

````python
fish_target = [1] * 35 + [0] * 14 # 결과값 생성: 1 = 도미, 0 = 빙어

print(fish_target)
````

````python
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()
````

````python
kn.fit(fish_data, fish_target) # trainning -> fit()
kn.score(fish_data, fish_target) # model 평가(Accuracy)
````

kn.predict([[30,600]])

````python
print(kn._fit_X)
print(kn._y)
````

**KNeighborsClassifier클래스는 기본적으로 가까운 5개의 데이터를 참고한다. (K = 5)**
하지만 **n_neighbors** 매개변수를 통해 변경이 가능하다.

````python
kn49 = KNeighborsClassifier(n_neighbors = 49)

kn49.fit(fish_data, fish_target)
kn49.score(fish_data, fish_target)
````

accuracy가 1에서 0.71로 낮아진 걸로 보았을 때 k=5로 보는 것이 도미와 빙어를 분류하는데 좋은 모델이었던 것으로 보인다!

# 확인문제
5 <= k <= 49 중에 score < 1로 되는 k를 구하여라

````python
kn = KNeighborsClassifier()
kn.fit(fish_data, fish_target)

for n in range(5, 50):
  kn.n_neighbors = n # 인접 이웃 개수 설정
  score = kn.score(fish_data, fish_target) # 점수 개산
  if score < 1:
    print(n, score)
    break
````

````python
fish_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0,
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0,
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0, 9.8,
                10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
fish_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0,
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0,
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0, 6.7,
                7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]
````

````python
import numpy as np
np.column_stack(([1,2,3], [4,5,6])) # cbind같이 2개의 리스트를 나란히 붙임
````

````python
fish_data = np.column_stack((fish_length, fish_weight))
print(fish_data[:5])
````

````python
print(np.ones(5)) # 원하는 개수만큼 넣으면 1을 만들어 준다. -> 0을 만들고 싶다? np.zeros()
# fish_target = np.column_stack((np.ones(35), np.zeros(14))) 아니지..
fish_target = np.concatenate((np.ones(35), np.zeros(14))) # np.concatenate((a,b)) = c(a,b)
print(fish_target )
````

````python
from sklearn.model_selection import train_test_split # 훈련-테스트 세트 나누기

train_input, test_input, train_target, test_target = train_test_split(
    fish_data, fish_target, random_state = 42 # np.random.seed(42) = random_state(42)
) # train_test_split은 test data set을 25%로 만들어 낸다!

print(train_input.shape, test_input.shape) # shape = dim이다!! 잊지말자
print(train_target.shape, test_target.shape)
print(test_target) # 원 데이터와 비율이 달라짐
````

````python
train_input, test_input, train_target, test_target = train_test_split(
    fish_data, fish_target, random_state = 42, stratify = fish_target
) # stratify 매개변수에 타깃 데이터 전달하면 클래스 비율에 맞게 데이터를 나눠줌

print(test_target)
````

````python
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()
kn.fit(train_input, train_target)
kn.score(test_input, test_target)

print(kn.predict([[25, 150]])) # 좀 이상하게 분류되는 거 같은 데이터
````

````python
import matplotlib.pyplot as plt
plt.scatter(train_input[:,0],train_input[:,1])
plt.scatter(25,150, marker = '^') # marker는 표시모양을 설정, marker = '^'는 삼각형
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
````

````python
distance, indexes = kn.kneighbors([[25,150]]) # default: n_neighbors = 5, 5개의 이웃값을 반환
print(distance)
print(indexes)
````

````python
plt.scatter(train_input[:,0],train_input[:,1])
plt.scatter(25,150, marker = '^')
plt.scatter(train_input[indexes,0], train_input[indexes, 1], marker = 'D') # marker = 'D'는 마름모
plt.show() # 대부분의 값은 빙어
print(train_target[indexes])
````

위 그래프를 보면 x축와 y축의 스케일이 다른 것을 확인할 수 있다.

````python
plt.scatter(train_input[:,0],train_input[:,1])
plt.scatter(25,150, marker = '^')
plt.scatter(train_input[indexes,0], train_input[indexes, 1], marker = 'D')
plt.xlim((0,1000))
plt.show()
````

````python
mean = np.mean(train_input, axis = 0)
std = np.std(train_input, axis = 0)
print(mean,std)

train_scaled = (train_input - mean)/ std # 브로드캐스팅

print(train_scaled[:5])
````

````python
plt.scatter(train_scaled[:,0],train_scaled[:,1])
plt.scatter(25,150, marker = '^')
plt.show() # train data만 표준화했으니, sample data는 outlier처럼 표기
````

````python
new = ([25,150] - mean) / std
plt.scatter(train_scaled[:,0],train_scaled[:,1])
plt.scatter(new[0], new[1], marker = '^')
plt.show()
````

````python
kn.fit(train_scaled, train_target)
````

````python
test_scaled = (test_input - mean) / std
kn.score(test_scaled, test_target)
````
